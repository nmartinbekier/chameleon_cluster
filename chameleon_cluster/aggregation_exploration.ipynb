Ideas:
* Order each iterations merge score and plot
* Plot only the values of the best merge scores of the iteration
* Print also how many clusters are there currently, and how many merge score values (There are 'x' merge scores between these clusters)

```python
def plot_merge_scores(scores):
    x_val = range(len(scores))
    plt.plot(x_val, scores)
```

```python
plt.figure(figsize = (6, 6))
plot_merge_scores(merge_scores[-10])
```
