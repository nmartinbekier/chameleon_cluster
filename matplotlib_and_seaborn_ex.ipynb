```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generating data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a line plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(x, y, '-r', label='Sine wave')
plt.title('Line Plot with Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Creating a distribution plot using Seaborn
# For the distribution plot, let's generate some random data
data = np.random.normal(size=1000)

plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='blue', label='Data Distribution')
plt.title('Distribution Plot with Seaborn')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
